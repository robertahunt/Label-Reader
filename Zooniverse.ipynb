```python
from collections import Counter

threshold = 80

useful_columns = ['coll', 'type', 'loc', 'day', 'month', 'year', 'collector', 
       'name', 'sex', 'ZMUC', 'determinator', 'preparator',
       'genital_num','person','subject_ids']
df2 = df[useful_columns].pivot(columns='subject_ids',index='person').T
df2 = df2.replace('Not specified (unknown)',np.nan).replace('[unclear][/unclear]',np.nan).replace('[uklar][/uklar]',np.nan).replace(False, np.nan)
df2 = df2.replace('',np.nan).replace(' ',np.nan).dropna(how='all',axis=0)

df2['#agree_exact'] = np.nan
df2['best_exact'] = np.nan
df2['#agree_fuzz'] = np.nan
df2['best_fuzz'] = np.nan
for name, row in tqdm(df2.iterrows(), total=len(df2)):
    row = row.dropna()
    counts = row.groupby(row).count().sort_values(ascending=False)
    df2.loc[name,'#agree_exact'] = counts.values[0]
    df2.loc[name,'best_exact'] = counts.index[0]
    ratios = []
    if len(row) == 1:
        df2.loc[name, 'best_fuzz'] = row.iloc[0]
        df2.loc[name, '#agree_fuzz'] = 1
    else:
        for i in range(len(row)):
            for j in range(i+1,len(row)):
                ratio = fuzz.token_sort_ratio(row.iloc[i],row.iloc[j])
                ratios += [[i,ratio],[j,ratio]]
        ratios = pd.DataFrame(ratios,columns=['index','ratio'])
        best = ratios.groupby('index').mean().sort_values('ratio',ascending=False).index[0]
        agree = max(len(np.unique(ratios[ratios['ratio']>=threshold].index)), 1)
        df2.loc[name, 'best_fuzz'] = row.iloc[best]
        df2.loc[name, '#agree_fuzz'] = agree
```
