```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
```

```python
train = pd.read_csv('/home/rob/HWR/SimpleHTRLatest/SimpleHTR/model_from_monday/train_losses.txt',sep=' ',names=['epoch','loss'])
```

```python
val = pd.read_csv('/home/rob/HWR/SimpleHTRLatest/SimpleHTR/model_from_monday/val_losses.txt',sep=' ',names=['loss','charErrorRate','WordErrRate'])
val['WordErrRate'] = 1-val['WordErrRate']

plt.figure()
plt.plot(train['loss'],label='Train Set')
plt.plot(val['loss'], label='Validation Set')
plt.title('Losses during training, SimpleHTR Model')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
```

```python
plt.figure()
plt.plot(val['charErrorRate'],label='CER')
plt.plot(val['WordErrRate'],label='WER')
plt.title('Validation Set CER and WER during training, SimpleHTR Model')
plt.xlabel('epoch')
plt.ylabel('CER')
plt.legend()
```
